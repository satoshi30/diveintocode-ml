## 実装していく前に

ニューラルネットワークは一般化線形モデルの多層化と同値です。
一般化線形モデルとは、今までにやって来た線形回帰やロジスティック分類を総称したものです。数式上では  
$$
y=g(\beta_0 + f_1(x_1) + f_2(x_2) + f_3(x_3))
$$  
と表わすことができるモデルのことです。一般にこれは一般化加法モデルとも呼ばれます。関数$f_i$を特徴量抽出の関数だとみなすことができますが、深層学習では中間層がこの特徴量抽出器になっており、線形和とバイアスの出力を重ねることでこれを実現していると解釈することができるので、
$$
y=f(\beta_0 + ax_1 + bx_2 + cx_3)
$$  
という簡単なモデルまで落とし込めます。
上式の$f$がsigmoidやtanh、さらにはReLUになっており、出力を複数に分岐させ、それを次の層の入力（特徴量）とするモデルをニューラルネットワークといいます。

**出力を分岐**  
線形回帰モデルではa,b,c等を係数にして、一次元行列としていましたが、これを二次元行列にすることで出力を複数に分けることができます。以下に入力次元n、出力次元がmとなるモデルを書きます。

出力$y$、入力$X$、重み$W$、バイアスは以下

$$
\
y = \left[
  \begin{array}{ccc}
    y_1 \\
    y_2 \\
    \vdots \\
    y_m
  \end{array}
    \right]
\
,
\
W = \left[
  \begin{array}{ccc}
    w_{11} & w_{12} & \cdots & w_{1n} \\
    w_{21} & w_{22} & \cdots & w_{2n} \\
    \vdots & \vdots & \vdots & \vdots \\
    w_{m1} & w_{m2} & \cdots & w_{mn}
  \end{array}
    \right]
\
,  \\
\
X = \left[
  \begin{array}{ccc}
    x_{1} \\
    x_{2} \\
    \vdots \\
    x_{n}
  \end{array}
    \right]
\
,
\
b = \left[
  \begin{array}{ccc}
    b_{1} \\
    b_{2} \\
    \vdots \\
    b_{m}
  \end{array}
    \right]
\
$$

で表すことができ、ある層の入力と出力の関係式は$y=WX+b$で記述できます。バッチ学習する際はバッチサイズも考慮する場合には注意する必要がありますが、基本的には$y$と$X$の次元が変わるだけです。

以上を踏まえて実装していきましょう。

# 【問題４】back propagationの実装


バックプロパゲーションの実装(問題４)を行う前に微分や連鎖率についてまとめなおしておきましょう。  

まずは出力層を考えます。出力層のノード $k$ への入力の総和を $\lambda_k$ とすると出力層では出力が

\begin{align}
y_k = \frac{\exp{(\lambda_k)}}{\sum_{k=1}^{K}\exp{(\lambda_k)}}
\end{align}

となっているはずです。
ひとつ前の層のノード数を $n$ とすると、

\begin{align}
\lambda_k = \sum_{i=1}^{n}w_{ki}x_{i}+b_k
\end{align}


損失関数 $E$ からパラメータを更新したいので、偏微分の連鎖律を考えると、

\begin{align}
\frac{\partial E}{\partial w_{ki}}=\frac{\partial E}{\partial \lambda_k} \frac{\partial \lambda_k}{\partial w_{ki}}
\end{align}
となります。

自然に$\frac{\partial E}{\partial \lambda_k}$と$\frac{\partial \lambda_k}{\partial w_{ki}}$を導出すればいいことになります。
それぞれの微分を考えていきます。


\begin{align}
\frac{\partial E}{\partial \lambda_k}
&= \frac{\partial}{\partial \lambda_k}\Bigr(-\sum^{K}_{k=1} t_k \times \log(y_k)\Bigl)\\
&= \frac{\partial}{\partial \lambda_k}\Bigr(-\sum^{K}_{k=1} t_k \log \bigl(\frac{\exp{(\lambda_k)}}{\sum_{k=1}^{K}\exp{(\lambda_k)}}\bigr)\Bigl)  \\
&= \frac{\partial}{\partial \lambda_k}\Bigr(-\sum^{K}_{k=1} t_k \bigl( \lambda_k - \log \sum_{k=1}^{K} exp(\lambda_k) \bigr)\Bigr)  \\
&= \frac{\partial}{\partial \lambda_k}\Bigr(-\sum^{K}_{k=1} t_k \lambda_k + \log \sum_{k=1}^{K} exp(\lambda_k) \Bigr)  \\
&= y_k - t_k
\end{align}



となります。ただし、
\begin{align}　
t_i = 
\begin{cases}
    1  ,(正解時)  \\
    0  ,(不正解時)
\end{cases}
\end{align}
です。

\begin{align}
\frac{\partial \lambda_k}{\partial w_{ki}} &= \frac{\partial}{\partial w_{ki}}\Bigl(\sum_{i=1}^{n} w_{ki}x_i + b\Bigr)  \\
&= x_i
\end{align}

従って、パラメータの更新は

\begin{align}
w_{ki} = w_{ki} - \alpha(y_k - t_k)x_i
\end{align}
すなわち、全てのパラメータは
\begin{align}
w = w - \alpha(y-t)x^T
\end{align}
で行うことができます。この時の行列の形は $y-t$ が(K,batch), $x$ が(n,batch)となっています。

ほぼ同様の手順でバイアス項の学習も行えます。


以上で出力層のパラメータ更新に必要な式は終わりですが、中間層で使う出力層における入力の勾配をここで求める必要があります。

\begin{align}
\frac{\partial E}{\partial x_{i}}
&= \sum_{k=1}^{K} \frac{\partial E}{\partial \lambda_k} \frac{\partial \lambda_k}{\partial x_i}  \\
&= \sum_{k=1}^{K} (y_k - t_k) \frac{\partial \lambda_k}{\partial x_i}  \\
&= \sum_{k=1}^{K} (y_k - t_k) w_{ki}
\end{align}


**中間層**についても考えて活きます。中間層のノード $i$ への入力の総和を $\mu_i$ 、一つ前のノード数を $n$ とすると

\begin{align}
\mu_i = \sum_{j=1}^{n}w_{ij}x_{j}+b_i
\end{align}

となります。エントロピーから$w$を更新したいので、
\begin{align}
\frac{\partial E}{\partial w_{ij}}
&= \frac{\partial E}{\partial \mu_i} \frac{\partial \mu_i}{\partial w_{ij}}  \\
&= \frac{\partial E}{\partial y_i} \frac{\partial y_i}{\partial \mu_i} \frac{\partial \mu_i}{\partial w_{ij}}
\end{align}

\begin{align}
\frac{\partial E}{\partial y_i} &： 出力層の入力の勾配  \\
\frac{\partial y_i}{\partial \mu_i} &： 活性化関数の微分
\end{align}

この時 $y_i$ はノード$i$の出力です。つまり出力層の入力です。これにより、$\frac{\partial E}{\partial w_{ij}}$ が求まりました。

また、さらに中間層が続く場合は、$l+1$ 層目の入力の勾配を出力層の勾配と同じように計算すると上の層である $l$ 層目の勾配が求まります。
