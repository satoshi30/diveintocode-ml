## 【問題】栗まんじゅうが太陽系を覆う日
1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。


栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。


ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。

```python
#球状の物体として体積を計算
def volume(radius):
    import numpy as np
    volume = 3 * np.pi / 4 * radius**3
    return volume
```

```python
light_year =9_460_730_472_580_800               #1光年をメートルに変換
solar_system_radius = 45 * 10**8 * light_year     # 太陽から海王星までの距離を太陽系の半径と仮定、45億光年をmに変換
kurimanjyu_radius = 3 / 100                                     #　栗まんじゅうの半径:3cmと仮定：mに変換
```

```python
solar_system_volume = volume(solar_system_radius)
kurimanjyu_volume = volume(kurimanjyu_radius)
```

```python
def baibain(kurimanjyu_volume, another_volume, another_name):
    num = 0
    baibain_volume = kurimanjyu_volume
    kurimanjyu_lst = [baibain_volume]
    elapsed_minute_lst = [num]
    while baibain_volume < another_volume:
        baibain_volume *= 2
        num += 1
        kurimanjyu_lst.append(baibain_volume)
        elapsed_minute_lst.append(num * 5)
    print('{}分後に、{}が栗まんじゅうで一杯になる'.format(elapsed_minute_lst[-1], another_name))
    import matplotlib.pyplot as plt
    %matplotlib inline
    plt.plot(elapsed_minute_lst, kurimanjyu_lst, label='kurimanjyu_volume', marker='.', linewidth=0)
    plt.hlines(another_volume, elapsed_minute_lst[0], elapsed_minute_lst[-1], label=another_name+'_volume', color='red', ls='--')
    plt.title('baibain')
    plt.xlabel('elapsed_minute')
    plt.ylabel('volume[m3]')
    plt.legend();
```

```python
baibain(kurimanjyu_volume, solar_system_volume, 'solar_system')
```

```python
tokyo_dome_volume = 124 * 10**4      # https://www.tokyo-dome.co.jp/faq/dome/ 容積は124万立方メートルです。
baibain(kurimanjyu_volume, tokyo_dome_volume, 'tokyo_dome')
```
