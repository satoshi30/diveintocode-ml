yolo_video.pyの関数、detect_img(yolo)の中にコードを追加する
```
def detect_img(yolo):
    while True:
        img = input('Input image filename:')
        try:
            image = Image.open(img)
        except:
            print('Open Error! Try again!')
            continue
        else:
            r_image = yolo.detect_image(image)
            r_image.show()
            
            # 画像保存のためコード追加
            import cv2
            import numpy as np
            from datetime import datetime
            out_time = "{0:%Y%m%d-%H%M%S}".format(datetime.now())
            out_path = "output_image/" + out_time + ".jpg"
            cv2.imwrite(out_path, np.asarray(r_image)[..., ::-1])

    yolo.close_session()
```
